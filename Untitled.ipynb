```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

from keras.layers import LSTM,Dropout,SimpleRNN,Dense
from keras.models import Sequential
```

```python
df = pd.read_csv('Uniqlo(FastRetailing) 2012-2016 Training - stocks2012-2016.csv')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.set_index('Date',inplace=True)
```

```python
df['Close'].plot()
```

```python
df['Volume'].plot()
```

```python
train = df[:len(df)-226]
valid = df[len(train):]
```

```python
ft = train.drop('Close',axis=1)
tt = train[['Close']]
```

```python
ft.head()
```

```python
tt.head()
```

```python
sc = MinMaxScaler(feature_range=(0,1))
train_scalled = sc.fit_transform(ft)
train_scalled
```

```python
st = MinMaxScaler(feature_range=(0,1))
tt_scalled = st.fit_transform(tt)
tt_scalled
```

```python
X_train , y_train = [],[]
timesteps = 50

for i in range(timesteps,len(train_scalled)):
    X_train.append(train_scalled[i-timesteps:i,:])
    y_train.append(tt_scalled[i,0])
```

```python
X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
X_train = X_train.reshape(X_train.shape[0],X_train.shape[1],5)
```

```python
X_Val = df[len(df) - len(valid) - timesteps:]
X_Val.drop('Close',axis=1,inplace=True)
```

```python
XV_scalled = sc.transform(X_Val)
```

```python
X_VAL = []
for i in range(timesteps,len(XV_scalled)):
    X_VAL.append(XV_scalled[i-timesteps:i,:])
```

```python
X_VAL = np.array(X_VAL)
X_VAL.shape
```

```python
tval = st.transform(valid[['Close']])
```

```python
model = Sequential()
model.add(LSTM(16,activation='tanh',input_shape=(X_train.shape[1],5),return_sequences=False))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(X_train,y_train,batch_size=16,epochs=100)
```

```python
prediction = model.predict(X_VAL)
```

```python
nprediction = st.inverse_transform(prediction)
```

```python
ntval = st.inverse_transform(tval)
```

```python
plt.figure(figsize=(10,7))
plt.plot(ntval,label = 'Real Value')
plt.plot(nprediction, label='Prediction')
plt.legend()
plt.grid()
plt.show()
```
